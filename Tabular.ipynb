```python
import numpy as np
import pandas as pd
```

```python
# Training set
train=pd.read_csv('train.csv')
train.head(5)
```

```python
# Testing set
test=pd.read_csv('test.csv')
test.head(5)
```

```python
# Splitting of datasets.
x_train,y_train=train.drop(columns=['failure']),train['failure']
x_test=test.copy()
```

```python
# Drop
x_train.drop(columns=['id'],inplace=True)
x_test.drop(columns=['id'],inplace=True)
```

```python
# data preprocessing
col=['attribute_0', 'attribute_1']
```

```python
# Encoding
from feature_engine.encoding import WoEEncoder,RareLabelEncoder
re=RareLabelEncoder(variables=list(col))
re.fit_transform(x_train)
re.transform(x_train)
re.transform(x_test)
we=WoEEncoder(variables=list(col))
we.fit(x_train,y_train)
# Transform
x_train=we.transform(x_train)
x_test=we.transform(x_test)
```

```python
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
oe=LabelEncoder()
x_train['product_code']=oe.fit_transform(x_train['product_code'])
```

```python
x_test['product_code']=oe.fit_transform(x_test['product_code'])
```

```python
# data preprocessing
train.isnull().sum()/len(train)*100
```

```python
# Handling of null values
from sklearn.impute import KNNImputer
knn_null=KNNImputer()
knn_null.fit_transform(x_train)
```

```python
# Dealing with null values
x_train=pd.DataFrame(data=knn_null.fit_transform(x_train),columns=x_train.columns)
```

```python
x_test=pd.DataFrame(data=knn_null.transform(x_test),columns=x_test.columns)
```

```python
col=['loading', 'attribute_2',
       'attribute_3', 'measurement_0', 'measurement_1', 'measurement_2',
       'measurement_3', 'measurement_4', 'measurement_5', 'measurement_6',
       'measurement_7', 'measurement_8', 'measurement_9', 'measurement_10',
       'measurement_11', 'measurement_12', 'measurement_13', 'measurement_14',
       'measurement_15', 'measurement_16', 'measurement_17']
```

```python
# Transformer
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
for i in col:
    x_train[i]=pt.fit_transform(x_train[[i]])
    x_test[i]=pt.transform(x_test[[i]])
```

```python
# Prepare for Model building
x_train.head(5)
x_test.head(5)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PowerTransformer
from sklearn.feature_selection import RFE,SelectFromModel
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# Model builidng.
linreg=LogisticRegression()
linreg.fit(x_train,y_train)
```

```python
# Predict
y_pred=linreg.predict(x_test)
```
