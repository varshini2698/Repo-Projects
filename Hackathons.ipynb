```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
# Training
train=pd.read_csv('train.csv')
train.head(3)
```

```python
# Testing 
test=pd.read_csv('test.csv')
test.head(3)
```

```python
# Split of data
x_train,y_train=train.drop(columns=['income']),train['income']
x_test=test.copy()
```

```python
# Check for Null values
d=x_train.isnull().sum()/len(x_train)*100
```

```python
d.values[d.values>1]
```

```python
x_test.isnull().sum()/len(x_test)*100
```

```python
# Dropping of Column
x_train.drop(columns=['currently_enrolled','labour_union','major_occ_code','house_1yr_ago',
                      'income_filled','under18','unemp_reason','woker_type'],inplace=True)
```

```python
# Dropping of Column
x_test.drop(columns=['currently_enrolled','labour_union','major_occ_code','house_1yr_ago',
                      'income_filled','under18','unemp_reason','woker_type'],inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
```

```python
dummy=[]
ord1=[]
for i in x_train.describe(include='object').columns:
    if x_train[i].nunique()>3:
        ord1.append(i) 
    else:
        dummy.append(i)
```

```python
dummy
```

```python
x_train=pd.get_dummies(columns=['gender'],data=x_train,drop_first=True)
x_test=pd.get_dummies(columns=['gender'],data=x_test,drop_first=True)
```

```python
x_train
```

```python
# Fill the null values
from sklearn.impute import KNNImputer
knni=KNNImputer()
```

```python
knni.fit(x_train,y_train)
```

```python
# Null values

x_train=pd.DataFrame(data=knni.transform(x_train),columns=x_train.columns)
x_test=pd.DataFrame(data=knni.transform(x_test),columns=x_test.columns)
```

```python
y_train.replace({'Less than 50K $':0,'More than 50K $':1},inplace=True)
```

```python
x_train.drop(columns=['ID'],inplace=True)
x_test.drop(columns=['ID'],inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
logit=LogisticRegression()
```

```python
logit.fit(x_train,y_train)
```

```python
y_pred=logit.predict(x_test)
```

```python
logit.score(x_train,y_train)
```

```python
from sklearn.metrics import roc_auc_score
```
